```python
import pydicom
import matplotlib.pyplot as plt

# Charger le DICOM
dicom_path = '1_IM-0001-3001.dcm'
ds = pydicom.dcmread('1_IM-0001-3001.dcm')

# Afficher les métadonnées
print(ds)

# Afficher l'image
plt.figure(figsize=(10, 10))
plt.imshow(ds.pixel_array, cmap='gray')
plt.title('DICOM Image')
plt.colorbar()
plt.show()
```
