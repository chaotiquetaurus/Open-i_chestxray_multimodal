# 📊 Artishow PNG Dataset Scanner - Google Colab Edition

**Objectif:** Scanner le dataset PNG sur Google Drive et exporter la liste des fichiers pour mapping local

**Environnement:** Google Colab + Google Drive

**Output:**
- ✅ `png_files_list.csv` - Liste complète des PNG avec noms
- ✅ Statistiques: nombre de fichiers, répertoire scannée

## 🔌 Section 1: Authentification Google Drive

```python
# Authentifier et monter Google Drive
from google.colab import drive
drive.mount('/content/drive')

print("✅ Google Drive monté avec succès!")
```

## 📁 Section 2: Configuration et Scan du Dataset

```python
import os
import pandas as pd
from datetime import datetime

# Configuration des chemins sur Google Drive
DRIVE_BASE = '/content/drive/MyDrive'
ARTISHOW_DIR = os.path.join(DRIVE_BASE, 'artishow')
PNG_DIR = os.path.join(ARTISHOW_DIR, 'preprocessed_images_multiwindow')
OUTPUT_DIR = os.path.join(ARTISHOW_DIR, 'dataset_analysis_colab')

# Créer le dossier de sortie
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f"{'='*60}")
print("📁 CONFIGURATION DES RÉPERTOIRES")
print(f"{'='*60}")
print(f"\n📂 Répertoire PNG: {PNG_DIR}")
print(f"📂 Dossier output: {OUTPUT_DIR}")

# Vérifier que le répertoire PNG existe et scanner
if os.path.exists(PNG_DIR):
    all_files = os.listdir(PNG_DIR)
    png_files = [f for f in all_files if f.lower().endswith('.png')]
    print(f"\n✅ Répertoire trouvé!")
    print(f"   Fichiers totaux: {len(all_files)}")
    print(f"   Fichiers PNG: {len(png_files)}")
else:
    print(f"\n❌ Répertoire PNG non trouvé!")
    png_files = []
```

```python
print(f"\n{'='*60}")
print("📋 LISTE DES FICHIERS PNG")
print(f"{'='*60}\n")

# Créer un DataFrame avec les PNG
df_png = pd.DataFrame({
    'filename': png_files
})

# Ajouter un index
df_png.insert(0, 'id', range(1, len(df_png) + 1))

print(f"Total: {len(df_png)} fichiers PNG\n")
print(df_png.to_string(index=False))

# Sauvegarder en CSV
output_csv = os.path.join(OUTPUT_DIR, 'png_files_list.csv')
df_png.to_csv(output_csv, index=False)
print(f"\n✅ Fichier CSV généré: {output_csv}")
```

## ✅ Section 3: Résumé et Prochaines Étapes

```python
print(f"\n{'='*60}")
print("📊 RÉSUMÉ DU SCAN")
print(f"{'='*60}\n")

print(f"  ✅ Fichier PNG scannés: {len(df_png)}")
print(f"  📁 Répertoire: {PNG_DIR}")
print(f"  💾 CSV généré: png_files_list.csv")
print(f"\n📝 Prochaines étapes:")
print(f"  1. Télécharger 'png_files_list.csv' depuis Google Drive")
print(f"  2. Utiliser la liste en local pour créer le mapping")
print(f"  3. Re-uploader le CSV de mapping si nécessaire")

print(f"\n{'='*60}\n")
```
